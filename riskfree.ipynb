```python
prices = pd.read_csv("equity/data.csv", index_col="date")
prices=prices.iloc[::-1]
prices.index = pd.to_datetime(prices.index)
prices.head()
```

```python
cov_matrix = prices.resample('Y').last().pct_change().apply(lambda x: np.log(1+x)).cov()
cov_matrix
```

```python
e_r = prices.pct_change().mean()
e_r
```

```python
sd = prices.pct_change().apply(lambda x: np.log(1+x)).std().apply(lambda x: x*np.sqrt(250))
sd
```

```python
assets = pd.concat([e_r, sd], axis=1)
assets.columns = ['Returns', 'Volatility']
assets
```

```python
p_ret = []
p_vol = []
p_weights = []

num_assets = len(prices.columns)
num_portfolios = 1000
```

```python
for portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights = weights/np.sum(weights)
    p_weights.append(weights)
    returns = np.dot(weights, e_r)
    p_ret.append(returns)
    var = cov_matrix.mul(weights, axis=0).mul(weights, axis=1).sum().sum()
    sd = np.sqrt(var)
    ann_sd = sd*np.sqrt(250)
    p_vol.append(ann_sd)
```

```python
data = {'Returns':p_ret, 'Volatility':p_vol}
```

```python
for counter, symbol in enumerate(prices.columns.tolist()):
    data[symbol+' weight'] = [w[counter] for w in p_weights]
```

```python
portfolios = pd.DataFrame(data)
portfolios.head()
```

```python
from decimal import Decimal


def lin_equ(l1, l2):
    """Line encoded as l=(x,y)."""
    m = (l2[1] - l1[1]) / l2[0] - l1[0]
    c = (l2[1] - (m * l2[0]))
    return m, c
```
